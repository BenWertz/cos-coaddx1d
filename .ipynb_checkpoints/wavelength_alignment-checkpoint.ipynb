```python
import numpy as np
from scipy.interpolate import *
from astropy.stats import sigma_clip
import astropy.units as u
import astropy.constants as c
import astropy.io.fits as fits
import matplotlib.pyplot as plt
from astropy.modeling import *
from scipy import signal
from scipy.optimize import curve_fit
from scipy.ndimage import gaussian_filter1d
from linetools.lists.linelist import LineList
```

```python
llist_ism = LineList('ISM')._data.to_pandas()

line_list=["SII 1250","SII 1253","AlII 1670"]


llist_ism['species'] = llist_ism['name'].str.split().str[0]

sub_llist = llist_ism.query('(name in @line_list)').sort_values("f") # from low to high
display(sub_llist)

line_lib = {"line_name": [line.replace(' ', '') for line in sub_llist['name']],
             "lambda": list(sub_llist['wrest']*u.AA), 
             "f": list(sub_llist['f'])}
line_lib
```

```python
# data=fits.open("../cos_data_reduction/targets/PG1011-040/raw_g160m/x1d/lb4q04a3q_x1d.fits")
data=fits.open("../cos_data_reduction/targets/PG1011-040/raw_g130m/x1d/lb4q04anq_x1d.fits")
spec_hdu=data[1]
spec_hdu.header
```

```python
wave=np.concatenate([
    spec_hdu.data["WAVELENGTH"][1,:],
    spec_hdu.data["WAVELENGTH"][0,:]
])*u.AA

flux=np.concatenate([
    spec_hdu.data["FLUX"][1,:],
    spec_hdu.data["FLUX"][0,:]
])

error=np.concatenate([
    spec_hdu.data["ERROR"][1,:],
    spec_hdu.data["ERROR"][0,:]
])

# dv=1#(c.c*wave.mean()/(wave[1]-wave[0])).to(u.km/u.s)
# high_pass=signal.butter(10,1/20,'high',fs=dv)
```

```python

# Gaussian absorption line shape
def gaussian_line_shape(v,a,v0,sigma_v):
    return 1-a*np.exp(-0.5*(v-v0)**2/sigma_v**2)
```

```python
linfit=fitting.LinearLSQFitter()

def continuum_normalize(mask,velocity,flux,error,curvefit_degree=2):
    if len(mask)==0:return flux,error,None
    vel_masked=velocity[mask]
    flux_masked=flux[mask]
    # Perform polynomial curve fit to approximate shape of continuum
    curvefit=linfit(
        polynomial.Polynomial1D(degree=curvefit_degree),
        vel_masked,
        flux_masked
    )
    flux_normalized=flux/curvefit(velocity)
    error_normalized=error/curvefit(velocity)
    return flux_normalized,error_normalized,curvefit

# Velocities lower than this are masked out when doing continuum curve fit
line_threshold=(100*u.km/u.s)

for name,line,f in zip(line_lib["line_name"],line_lib["lambda"],line_lib["f"]):
    if line<wave.min() or line>wave.max():continue
    # c*(l-l0)/l0
    vlsr=(c.c*(wave/line-1)).to(u.km/u.s)
    mask=(np.abs(vlsr)<(400*u.km/u.s))

    line_vlsr=vlsr[mask]
    line_flux=flux[mask]
    line_error=error[mask]

    line_flux_norm,line_error_norm,cont_curvefit=continuum_normalize(
        (np.abs(line_vlsr)>line_threshold),
        line_vlsr,
        line_flux,
        line_error
    )
    
    # Smoothing the signal out by a lot to identify spikes in the data
    flux_smoothed_1=gaussian_filter1d(line_flux_norm,6)
    
    # Comparing the actual flux to the smoothed version before running sigma clipping:
    # that way the code can remove random spikes in the data without also cutting out the absorption line.
    sigma_clip_mask=sigma_clip(
        line_flux_norm - flux_smoothed_1,
        sigma=2.5, # Note: is this is too low? too high?
        masked=True
    ).mask # only need the mask, since it needs to filter the velocity and flux arrays

    # Filter data based on the mask 
    vlsr_clipped_0=line_vlsr[~sigma_clip_mask]
    flux_clipped_0=np.maximum(line_flux_norm[~sigma_clip_mask],0)
    error_clipped_0=line_error_norm[~sigma_clip_mask]
    
    # Re-interpolate flux (and error) to patch holes in data
    line_flux_pc_0=interp1d(vlsr_clipped_0,flux_clipped_0,fill_value='extrapolate')(line_vlsr)
    line_error_pc=interp1d(vlsr_clipped_0,error_clipped_0,fill_value='extrapolate')(line_vlsr)
    
    line_flux_pc=line_flux_pc_0#gaussian_filter1d(line_flux_pc_0,1)
    
    params,cov=curve_fit(
        gaussian_line_shape,
        line_vlsr.value,
        line_flux_pc,
        p0=(
            1-line_flux_pc.min(), #amplitude
            0, #center velocity
            50/2.355 #stddev (TODO: calculate this from FWHM (how to do that??))
        ),
        bounds=(
            (0,-50,1), #(amplitude,v0,stddev)
            (1,50,300), #(amplitude,v0,stddev)
        )
    )
    
    # Reduced chi-squared value for Gaussian fit
    chisq_reduced=np.sum(
        (line_flux_pc-gaussian_line_shape(line_vlsr.value,*params))**2/line_error_pc**2
    )/(len(line_flux_pc)-3)
    print(params)
    print(cov)
    print("Chi_sq:",chisq_reduced)
    
    fig=plt.figure(figsize=(12,8))

    plt.plot(line_vlsr,line_flux_pc)
    plt.plot(
        line_vlsr,gaussian_line_shape(line_vlsr.value,*params)
    )

    plt.plot(line_vlsr,flux_smoothed_1)
    plt.scatter(line_vlsr[sigma_clip_mask],line_flux_norm[sigma_clip_mask],c="r",marker="x")

    plt.xlim(-400,400)
    plt.title(name)
    
    plt.gca().axvspan(-500, -line_threshold.value, alpha=0.5, color='red')
    plt.gca().axvspan(line_threshold.value, 500, alpha=0.5, color='red')

plt.show()
```
